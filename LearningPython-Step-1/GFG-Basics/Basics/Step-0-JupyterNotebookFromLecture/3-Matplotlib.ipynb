figure - area where I draw the graph (full container)

Axes - specific parts of the figure where individual plots are drawn

Axis - x axis and y axis

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,4))
x = [1,2,3,4]
y= [10,20,15,25]
ax1.plot(x,y)
ax1.set_xlabel("X values")
ax1.set_ylabel("Y values")
ax1.set_title("Left plot")

y2 = [2,4,5,6]
ax2.plot(x,y2)
ax2.set_xlabel("X values")
ax2.set_ylabel("Y values")
ax2.set_title("Right plot")

plt.show()
```

```python
fig, axes = plt.subplots(2,2, figsize=(10,10))

ax1, ax2, ax3, ax4 = axes.flatten()

x = [1,2,3,4]
y= [10,20,15,25]
ax1.plot(x,y)
ax1.set_xlabel("X values")
ax1.set_ylabel("Y values")
ax1.set_title("Left plot")

y2 = [2,4,5,6]
ax2.plot(x,y2)
ax2.set_xlabel("X values")
ax2.set_ylabel("Y values")
ax2.set_title("Right plot")

y3 = [1,4,5,0]
ax3.plot(x,y3)
ax3.set_xlabel("X values")
ax3.set_ylabel("Y values")
ax3.set_title("2nd left plot")

y4 = [2,9,11,6]
ax4.plot(x,y4)
ax4.set_xlabel("X values")
ax4.set_ylabel("Y values")
ax4.set_title("2nd Right plot")

plt.tight_layout()
plt.show()
```

IQR (Inter Quartile Range) =Q3-Q1\
then outliers???\
Lower bound = Q1 - 1.5*IQR\
Upper bound = Q3 + 1.5*IQR

```python
data1 = [10,12,14,15,15,16,18,19,100,120]
df1 = pd.DataFrame(data1,columns = ["Values"])
```

```python
Q1 = df1["Values"].quantile(.25)
Q3 = df1["Values"].quantile(.75)
```

```python
IQR = Q3-Q1
```

```python
Lower_bound = Q1 - 1.5*IQR
Upper_bound = Q3 + 1.5*IQR
```

```python
outliers = df1[(df1["Values"]<Lower_bound) | (df1["Values"]>Upper_bound )]
```

```python
outliers 
```

```python
plt.boxplot(df1["Values"])
plt.title("BoxPlot")
plt.show()
```
