IQR (Inter Quartile Range) =Q3-Q1\
then outliers???\
Lower bound = Q1 - 1.5*IQR\
Upper bound = Q3 + 1.5*IQR

```python
data1 = [10,12,14,15,15,16,18,19,100,120]
df1 = pd.DataFrame(data1,columns = ["Values"])
```

```python
Q1 = df1["Values"].quantile(.25)
Q3 = df1["Values"].quantile(.75)
```

```python
IQR = Q3-Q1
```

```python
Lower_bound = Q1 - 1.5*IQR
Upper_bound = Q3 + 1.5*IQR
```

```python
outliers = df1[(df1["Values"]<Lower_bound) | (df1["Values"]>Upper_bound )]
```

```python
outliers 
```

# Stacked plot

```python
days = np.arange(1,8)
apples = [1,2,3,1,3,4,5]
bananas = [2,4,1,5,6,1,1]
cherries = [1,2,3,1,0,1,2]
plt.stackplot(days,apples,bananas,cherries,labels = ["apples", "bananas", "cherries"],colors = ["red","yellow","green"])
plt.xlabel("Days of the week")
plt.ylabel("Number of fruits")
plt.title("Fruit Consumption")
plt.legend()
plt.show()
```

```python
days = np.arange(1,11)
category1 = np.random.randint(1,10,size =10)
category2 = np.random.randint(1,10,size =10)
category3 = np.random.randint(1,10,size =10)
plt.stackplot(days,category1, category2, category3,labels = ["category1", "category2", "category3"],colors = ["blue","orange","green"])
plt.xlabel("Categories")
plt.ylabel("Random values")
plt.title("Stack plot for random values")
plt.legend()
plt.show()
```

# 3D plotting

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection = "3d")

x = np.random.randint(1,10,size =10)
y= np.random.randint(1,10,size =10)
z=np.random.randint(1,10,size =10)

ax.scatter(x,y,z,color ="purple")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
ax.set_title("Simple 3d Scatter plot")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection = "3d")

x = np.random.rand(50)
y= np.random.rand(50)
z=np.random.rand(50)

ax.scatter(x,y,z,c = z,cmap = "plasma", s= 50)
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
ax.set_title("3d Scatter plot")
plt.show()
```

x and y are linear spaced 100 numbers from -5 to 5 and z = sin(x^2+y^2)

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection = "3d")

x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)
x, y  = np.meshgrid(x,y)
z = np.sin(x*x+y*y)

ax.plot_surface(x,y,z, cmap = "plasma")
```

# Heatmap

Correlation lies between -1 and 1\
corr =1 i.e., positively correlated\
corr =-1 i.e., negatively correlated\
corr = 0 i.e., not correlated

```python
data = pd.DataFrame({"Maths" : np.random.randint(50,100,100),"Science" : np.random.randint(40,90,100),"Social Science" : np.random.randint(30,90,100),"English" : np.random.randint(70,100,100)})
data
```

```python
corr_matrix = data.corr()
corr_matrix
```

```python
heatmap = plt.imshow(corr_matrix,vmin = -1, vmax=1)
plt.colorbar(heatmap, cmap = "coolwarm")
plt.show()
```
